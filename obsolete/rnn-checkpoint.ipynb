### Build model

```python
import tensorflow as tf

tf.reset_default_graph()

rnn_type = "lstm"
mode = "mi_hr_step_sleep"

num_features = Xall.shape[-1]
num_output = yall.shape[-1]
rnn_size = 8
state_size = sall.shape[-1]

global_step = tf.Variable(0, name='global_step', trainable=False)

# Placeholder
inputs = tf.placeholder(dtype=tf.float32,
                        shape=[None, None, num_features],
                        name="inputs")
targets = tf.placeholder(dtype=tf.float32,
                         shape=[None, None, num_output],
                         name="targets")

init_s = tf.placeholder(dtype=tf.float32,
                        shape=[None, state_size])

# Weight and bias initializers
initializer_weights = tf.contrib.layers.variance_scaling_initializer() #  #tf.contrib.layers.xavier_initializer_conv2d()
initializer_bias    = tf.constant_initializer(0.0) # 0

# Latent state encoder
with tf.variable_scope("latent_state_encoder", reuse=tf.AUTO_REUSE):
    encode_wc = tf.get_variable("encode_wc", shape=[state_size, rnn_size], initializer=initializer_weights)
    encode_bc = tf.get_variable("encode_bc", shape=[rnn_size], initializer=initializer_bias)
    
    encode_wh = tf.get_variable("encode_wh", shape=[state_size, rnn_size], initializer=initializer_weights)
    encode_bh = tf.get_variable("encode_bh", shape=[rnn_size], initializer=initializer_bias)
    
init_h = tf.tanh(tf.matmul(init_s, encode_wh) + encode_bh)
init_c = tf.tanh(tf.matmul(init_s, encode_wc) + encode_bc)

# Define RNN Cell
if rnn_type == "lstm":
    rnn = tf.contrib.rnn.BasicLSTMCell(rnn_size)
    # Initial state of the LSTM memory.
    initial_state = tf.nn.rnn_cell.LSTMStateTuple(init_c, init_h)
elif rnn_type == "gru":
    rnn = tf.contrib.rnn.GRUCell(rnn_size)
    # Initial state of the LSTM memory.
    initial_state = init_h
else:
    raise Exception('Unknown RNN type {}.'.format(rnn_type))
    
# Add dropout: will bring jitter???
# rnn = tf.nn.rnn_cell.DropoutWrapper(rnn, output_keep_prob=0.5)

# Dynamic rnn
latent_states, final_state = tf.nn.dynamic_rnn(rnn, inputs, initial_state=initial_state, dtype=tf.float32)
    
# Output layer
with tf.variable_scope("output_layer", reuse=tf.AUTO_REUSE):
    output_w = tf.get_variable("output_w", shape=[rnn_size, num_output], initializer=initializer_weights) 
    output_b = tf.get_variable("output_b", shape=[num_output], initializer=initializer_bias)

outputs = tf.matmul(tf.reshape(latent_states, [-1, rnn_size]), output_w) + output_b

# loss function: smooth L1 + L2/L1 reg
beta = 0.01
l2_reg = tf.nn.l2_loss(output_w) + tf.nn.l2_loss(encode_wc) + tf.nn.l2_loss(encode_wh)
#l1_reg = tf.contrib.layers.l1_regularizer(scale=0.005, scope=None)
#reg_loss = tf.contrib.layers.apply_regularization(l1_reg, [output_w])
reg_loss = tf.reduce_mean(beta * l2_reg)

# tf.reduce_sum(tf.losses.huber_loss(self.targets, self.predictions, delta=0.01, reduction=tf.losses.Reduction.NONE, loss_collection=None), axis=1)
# loss = tf.reduce_mean(tf.square(targets - outputs)) + tf.reduce_mean(beta * regularizer)
targets_flatten = tf.reshape(targets, [-1, num_output])
loss = tf.reduce_mean(tf.losses.huber_loss(targets_flatten, outputs, delta=1., reduction=tf.losses.Reduction.NONE, loss_collection=None))
loss += reg_loss # tf.reduce_mean(beta * regularizer)

# optimizer
learning_rate = tf.train.exponential_decay(1e-3, global_step, 1000, 0.95, staircase=True)
optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate)
grads_and_vars = optimizer.compute_gradients(loss)

# Perform gradient clipping by the global norm
grads, variables = zip(*grads_and_vars)
grads_clipped, _ = tf.clip_by_global_norm(grads, clip_norm=10.0)

# Apply the gradients after clipping them
train_op = optimizer.apply_gradients(zip(grads_clipped, variables),global_step=global_step)

# Training summaries
training_summaries = [
    tf.summary.scalar("train/loss", loss),
    tf.summary.scalar("train/learning_rate", learning_rate),
]

merged = tf.summary.merge(training_summaries)

# Checkpoint directory. Tensorflow assumes this directory already exists so we need to create it
root_dir = "_".join([rnn_type, pat, mode])
checkpoint_dir = os.path.abspath(os.path.join(root_dir, "checkpoints"))
checkpoint_prefix = os.path.join(checkpoint_dir, "model")
if not os.path.exists(checkpoint_dir):
    os.makedirs(checkpoint_dir)

# Create a saver for all variables
tf_vars_to_save = tf.trainable_variables() + [global_step] # save all trainable variable and global_step
saver = tf.train.Saver(tf_vars_to_save, max_to_keep=5)

# Initializer
init = tf.global_variables_initializer()
```

### Train

```python
batch_size = 32

with tf.Session() as sess:
    # Initialize summary writer
    summary_out_dir = os.path.join(root_dir, "summaries")
    summary_writer  = tf.summary.FileWriter(summary_out_dir, sess.graph)

    sess.run(init)

    max_iter = 2000
    
    best_val_loss = np.inf
    for t in range(max_iter):
        # Training
        train_all_loss = []
        for bid in range(0, Xtrain.shape[0]-batch_size, batch_size):
            train_feed_dict = { inputs: Xtrain[bid:bid+batch_size], 
                                targets: ytrain[bid:bid+batch_size], 
                                init_s: strain[bid:bid+batch_size] }

            train_step, train_summary, _, train_loss = sess.run([global_step, merged, train_op, loss], 
                                                                feed_dict = train_feed_dict)
            train_all_loss.append(train_loss)
        
        # Validation
        val_all_loss = []
        for bid in range(0, Xval.shape[0]-batch_size, batch_size):
            val_feed_dict = { inputs: Xval[bid:bid+batch_size], 
                              targets: yval[bid:bid+batch_size], 
                              init_s: sval[bid:bid+batch_size] }
            
            val_predictions, val_loss = sess.run([outputs, loss], feed_dict = val_feed_dict)
            val_all_loss.append(val_loss)
        
        # Save training process
        summary_writer.add_summary(train_summary, train_step)
        
        # Output
        if (t+1) % 100 == 0:
            print("{}, {}/{}, train loss: {:4f}, val loss: {:4f}".format(train_step, t+1, max_iter, np.mean(train_all_loss), np.mean(val_all_loss)))
            
            if best_val_loss >= np.mean(val_all_loss):
                best_val_loss = np.mean(val_all_loss)
                print("------Update best val loss------")
                path = saver.save(sess, checkpoint_prefix, global_step=train_step)
                print("[%s] Saving TensorFlow model checkpoint to disk." % datetime.now().strftime("%Y-%m-%d %H:%M"))
```
